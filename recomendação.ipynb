# Importação das bibliotecas de Pipeline, Standard Scaler e PCA

Pipeline para automatizar sequencias de tratamentos

Standard Scaler para padronizar valores do dataset entre 0 e 1, por que se não o cluster pode priorizar valores maiores.

PCA para reduzir a dimensionalidade do dataset, o que é bom para o algoritmo do KMeans

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

# Gerando o Pipeline e colocando PCA E Standard Scaler dentro para automatizar

```python
SEED = 1224
np.random.seed(1224)

pca_pipeline = Pipeline([('scaler', StandardScaler()),
                        ('PCA', PCA(n_components=2, random_state=SEED))])

```

# Fit e transform no dataset

```python
genre_embedding_pca = pca_pipeline.fit_transform(dados_generos1)

projection = pd.DataFrame(columns=['x', 'y'], data = genre_embedding_pca)
projection.head()
```

# Importando o cluster e dando fit e predict

```python
from sklearn.cluster import KMeans

clustering_pca = KMeans(n_clusters = 10, random_state=SEED)

clustering_pca.fit(projection)

dados_generos['cluster_pca'] = clustering_pca.predict(projection)
projection['cluster_pca'] = clustering_pca.predict(projection)
```

```python
projection['generos'] = dados_generos['genres']

projection.head()
```

# Código para gerar gráfico dos clusters

```python
fig = px.scatter(projection, x = 'x', y = 'y', color='cluster_pca', hover_data=['x', 'y', 'generos'])
fig.show()
```

# Função que plota as 10 recomendações juntas

```python
def vizualize(name, url):
    plt.figure(figsize=(15, 10))
    columns = 5
    
    for i, u in enumerate(url):
        
        ax = plt.subplot(len(url) // columns + 1, columns, i + 1)
        
        image = io.imread(u)
        
        plt.imshow(image)
        
        ax.get_yaxis().set_visible(False)
        
        plt.xticks(color = 'w', fontsize = 0.1)
        
        plt.yticks(color = 'w', fontsize = 0.1)
        
        plt.xlabel(name[i], fontsize = 10)
        
        plt.tight_layout(h_pad = 0.7, w_pad = 0)
        
#         plt.subplots_adjust(wspace = None, hspace = None)
        
        plt.tick_params(bottom = False)
        
        plt.grid(visible = None)
        
    plt.show()
        
```

# Procedimento de recomendação em uma só função

```python
def recomend(nome_musica):
    cluster = list(projection_m[projection_m['song']== nome_musica]['cluster_pca'])[0]
    musicas_recomendadas = projection_m[projection_m['cluster_pca'] == cluster][[0, 1, 'song']]

    x_musica = projection_m[projection_m['song'] == nome_musica][0][0]
    y_musica = projection_m[projection_m['song'] == nome_musica][1][0]


    distancias = euclidean_distances(musicas_recomendadas[[0, 1]], [[x_musica, y_musica]])
    musicas_recomendadas['id'] = dados['id']
    musicas_recomendadas['distancias'] = distancias
    recomendadas = musicas_recomendadas.sort_values('distancias').head(10)
    
    url = []
    name = []
    for i in recomendadas['id']:
        track = sp.track(i)
        name.append(track['name'])
        url.append(track['album']['images'][1]['url'])
    
    plt.figure(figsize=(15, 10))
    columns = 5
    
    for i, u in enumerate(url):
        
        ax = plt.subplot(len(url) // columns + 1, columns, i + 1)
        
        image = io.imread(u)
        
        plt.imshow(image)
        
        ax.get_yaxis().set_visible(False)
        
        plt.xticks(color = 'w', fontsize = 0.1)
        
        plt.yticks(color = 'w', fontsize = 0.1)
        
        plt.xlabel(name[i], fontsize = 10)
        
        plt.tight_layout(h_pad = 0.7, w_pad = 0)
        
#         plt.subplots_adjust(wspace = None, hspace = None)
        
        plt.tick_params(bottom = False)
        
        plt.grid(visible = None)
        
    plt.show()
    
```
